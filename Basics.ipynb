```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Salary_Data.csv')
df.head()
```

```python
df['YearsExperience'].head()
```

```python
x = df[['YearsExperience']]
y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(x_train,y_train)

lr.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = lr.predict(x_test)
mean_squared_error(y_pred, y_test,squared=False)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_data = df[['cut','clarity','color']]
ordinal_encoder = OrdinalEncoder(categories=[['Fair','Good','Very Good','Premium','Ideal'],
                                             ['I1','SI2','SI1','VS2','VS1','VVS2','VVS1','IF'],
                                             ['J','I','H','G','F','E','D']])
ordinal_encoded = ordinal_encoder.fit_transform(ordinal_data)
ordinal_df = pd.DataFrame(ordinal_encoded,columns=['cut_label','clarity_label','color_label'])
```

```python
ordinal_df
```

```python
pd.concat([ordinal_df,ordinal_data],axis=1)
```

```python
df_encoded = pd.concat([ordinal_df,df],axis=1)
df_encoded
#lr = LinearRegression()
```

```python
x  =df_encoded.drop(columns=['clarity','color','cut','price'])
y = df_encoded['price']
```

```python
x
```

```python
x_train,x_test,y_train,y_train = train_test_split(x,y,test_size=0.2,random_state=42)
```
